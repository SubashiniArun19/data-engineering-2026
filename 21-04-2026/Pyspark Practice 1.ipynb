```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("Practice PySpark").getOrCreate()
print("Created")
```

```python
data=[
    ("Subashini",95000,"AI"),
    ("Harini",90000,"Software"),
    ("Swaathihaa",92000,"Finance"),
    ("Varsha",88000,"Sales"),
    ("Srinithi",49000,"Software")
]
columns=["Name","Salary","Departments"]
df=spark.createDataFrame(data,columns)
df.show()
```

```python
df.select("Name","Salary").show()
```

```python
df.filter(df.Salary>50000).show()
```

```python
df.groupBy("Departments").count().show()
```

```python
df.orderBy("Salary").show()
```
