```python
empdf=spark.read.csv("employees.csv",header=True,inferSchema=True)
```

```python
from pyspark.sql.functions import to_date
empdf=empdf.withColumn("joining_date",to_date("joining_date","yyyy-MM-dd"))
empdf.show()
```

```python
from pyspark.sql.functions import year,month
empdf=empdf.withColumn("joining_year",year("joining_date"))\
            .withColumn("joining_month",month("joining_date"))
empdf.show()
```

```python
new_data = [
    (8, "Ravi", "IT", 68000, "2023-05-10", 2023, 5),
    (9, "Anita", "HR", 62000, "2023-06-15", 2023, 6)
]

columns = [
    "emp_id",
    "name",
    "department",
    "salary",
    "joining_date",
    "joining_year",
    "joining_month"
]

new_df = spark.createDataFrame(new_data, columns)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank
window_spec=Window.partitionBy("department").orderBy(col("salary").desc())
empdf=empdf.withColumn("rank",rank().over(window_spec))
empdf.show()
```

```python
from pyspark.sql.functions import row_number

empdf = empdf.withColumn("row_number", row_number().over(window_spec))
empdf.show()
```

```python
import ipywidgets as widgets
dept_dropdown=widgets.Dropdown(
    options=["IT","HR","Finance"],
    description="Department:"
)
display(dept_dropdown)
```

```python
salaryinslider=widgets.IntSlider(
    min=50000,
    max=80000,
    step=5000,
    description="Salary:"
)
display(salaryinslider)
```

```python
def filter_data(dept,sal):
  empdf.filter((empdf.department==dept)&(empdf.salary>=sal)).show()
  widgets.interact(filter_data,dept=dept_dropdown,salary=salaryinslider)
```

```python
top_emp=empdf.orderBy("salary",ascending=False).first()["name"]
```

```python
from pyspark.sql.functions import avg
avg_salary=empdf.select(avg("salary")).first()[0]
```

```python
dept_count=empdf.groupBy("department").count().collect()
```

```python
total_sales = saledf.select(sum("amount")).first()[0]
```

```python
best_sales_emp = saledf.groupBy("emp_id") \
    .sum("amount") \
    .orderBy("sum(amount)", ascending=False) \
    .first()["emp_id"]
```

```python
with open("final_report.txt", "w") as f:
    f.write(f"Top Employee: {top_emp}\n")
    f.write(f"Average Salary: {avg_salary}\n")
    f.write(f"Total Sales: {total_sales}\n")
    f.write(f"Best Sales Employee ID: {best_sales_emp}\n")
    f.write("Department Counts:\n")

    for row in dept_count:
        f.write(f"{row['department']}: {row['count']}\n")
```
