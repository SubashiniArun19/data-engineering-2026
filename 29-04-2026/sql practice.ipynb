```sql
create table employee_salary_analysis as

select
    emp_name,
    city,
    department,
    salary,
    experience,

    if(salary >= 95000, 'High',
        if(salary >= 80000, 'Medium', 'Low')
    ) as salary_category,

    avg(salary) over (partition by department) as department_avg_salary

from employees;
```
