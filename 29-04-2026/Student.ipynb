```python
from pyspark.sql.functions import col, avg, when
data = [
(101, "Arjun Kumar", "Python", 86, 92),
(102, "Neha Gupta", "Python", 74, 80),
(103, "Rahul Nair", "Data Engineering", 88, 85),
(104, "Priya Sharma", "AI", 91, 95),
(105, "Siddharth Rao", "Data Engineering", 67, 72),
(106, "Ayesha Khan", "AI", 79, 88),
(107, "Kavya Reddy", "Python", 94, 96),
(108, "Manish Yadav", "Data Engineering", 58, 69),
(109, "Divya Menon", "AI", 83, 90),
(110, "Farhan Ali", "Python", 62, 75)
]
columns = ["student_id", "student_name", "course", "marks", "attendance"]
df = spark.createDataFrame(data, columns)
display(df)
```

```python
df = df.withColumn("Grade",when(col("marks") >= 90, "A").when(col("marks") >= 75, "B").when(col("marks") >= 60, "C")
.otherwise("Needs Improvement"))
df.show()
```

```python
df = df.withColumn("Certification_status",when((col("marks") >= 75) & (col("attendance") >= 80), "Eligible").otherwise("Not Eligible"))
df.show()
```

```python
df_avg = df.groupBy("course").agg(avg("marks").alias("avg_percourse"))
df_avg.show()
```

```python
from pyspark.sql.functions import count
df_eligible = df.filter(col("certification_status") == "Eligible") .groupBy("course") .agg(count("*").alias("eligible_students"))
df_eligible.show()
```
