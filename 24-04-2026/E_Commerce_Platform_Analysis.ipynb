```python
spark.sql("""
SELECT
    c.customer_name,
    SUM(oi.quantity * p.product_price) AS total_revenue
FROM customers c
JOIN orders o
    ON c.customer_id = o.customer_id
JOIN order_items oi
    ON o.order_id = oi.order_id
JOIN products p
    ON oi.product_id = p.product_id
GROUP BY c.customer_name
ORDER BY total_revenue DESC
LIMIT 3
""").show()
```

```python
spark.sql("""
SELECT c.customer_name, SUM(oi.quantity * p.product_price) AS total_revenue
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id
JOIN order_items oi ON o.order_id = oi.order_id
JOIN products p ON oi.product_id = p.product_id
GROUP BY c.customer_name
ORDER BY total_revenue DESC
LIMIT 3
""").show()
```

```python
from pyspark.sql.functions import col

final_df = customers.alias("c") \
    .join(orders.alias("o"), col("c.customer_id") == col("o.customer_id")) \
    .join(order_items.alias("oi"), col("o.order_id") == col("oi.order_id")) \
    .join(products.alias("p"), col("oi.product_id") == col("p.product_id")) \
    .select(
        col("c.customer_id"),
        col("c.customer_name"),
        col("c.city"),
        col("p.product_name"),
        col("p.product_price"),
        col("oi.quantity"),
        (col("oi.quantity") * col("p.product_price")).alias("revenue")
    )

final_df.show()
```

```python
from pyspark.sql.functions import sum

top_customers = final_df.groupBy("customer_name") \
    .agg(sum("revenue").alias("total_revenue")) \
    .orderBy(col("total_revenue").desc())

top_customers.show(5)
```

```python
top_products = final_df.groupBy("product_name") \
    .agg(sum("revenue").alias("total_revenue")) \
    .orderBy(col("total_revenue").desc())

top_products.show(5)
```

```python
revenue_city = final_df.groupBy("city") \
    .agg(sum("revenue").alias("total_revenue")) \
    .orderBy(col("total_revenue").desc())

revenue_city.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank

windowSpec = Window.orderBy(col("total_revenue").desc())

ranked_customers = top_customers.withColumn(
    "rank", rank().over(windowSpec)
)

ranked_customers.show()
```
