```python
patients = spark.read.csv("patients.csv", header=True, inferSchema=True)
```

```python
from pyspark.sql.functions import when
patients = patients.withColumn("age_group",
    when(patients.age < 30, "Young")
    .when(patients.age < 50, "Adult")
    .otherwise("Senior")
)
```

```python
from pyspark.sql.functions import lit
patients = patients.withColumn("hospital_name", lit("BotCampus Hospital"))
```

```python
doctors = doctors.withColumn("fee_with_tax", doctors.consultation_fee * 1.18)
doctors.show()
```

```python
bills = bills.withColumn("bill_with_tax", bills.bill_amount * 1.18)
bills.show()
```

```python
bills = bills.withColumn("bill_in_thousands", bills.bill_amount / 1000)
bills.show()
```

```python
patients = patients.withColumn("country", lit("India"))
```

```python
from pyspark.sql.functions import concat, lit
doctors = doctors.withColumn(
    "doctor_label",
    concat("consultant_name", lit(" - "), "specialization")
)
doctors.show()
```

```python
patients = patients.withColumn(
    "patient_label",
    concat("name", lit(" - "), "city")
)
```

```python
bills = bills.withColumn(
    "high_bill_flag",
    when(bills.bill_amount > 1500, "Yes").otherwise("No")
)
```

```python
patients = patients.withColumn(
    "senior_patient_flag",
    when(patients.age > 40, "Yes").otherwise("No")
)
```

```python
from pyspark.sql.functions import when
patients = patients.withColumn(
    "age_category",
    when(patients.age < 30, "Young")
    .when(patients.age < 50, "Adult")
    .otherwise("Senior")
)
patients.show()
```

```python
bills = bills.withColumn(
    "bill_category",
    when(bills.bill_amount > 1500, "High")
    .when(bills.bill_amount >= 1000, "Medium")
    .otherwise("Low")
)
bills.show()
```

```python
doctors = doctors.withColumn(
    "fee_category",
    when(doctors.consultation_fee > 1500, "Premium")
    .when(doctors.consultation_fee >= 1000, "Standard")
    .otherwise("Basic")
)
doctors.show()
```

```python
appointments = appointments.withColumn(
    "priority",
    when(appointments.status == "Pending", "High")
    .when(appointments.status == "Completed", "Normal")
    .when(appointments.status == "Cancelled", "Low")
)
appointments.show()
```

```python
patients = patients.withColumn(
    "city_zone",
    when(patients.city.isin("Hyderabad", "Bangalore", "Chennai"), "South")
    .when(patients.city.isin("Mumbai", "Delhi", "Pune"), "Metro")
)
patients.show()
```

```python
from pyspark.sql.functions import to_date
patients = patients.withColumn(
    "registration_date",
    to_date("registration_date", "yyyy-MM-dd")
)
```

```python
from pyspark.sql.functions import year
patients = patients.withColumn("reg_year", year("registration_date"))
```

```python
from pyspark.sql.functions import month
patients = patients.withColumn("reg_month", month("registration_date"))
```

```python
appointments = appointments.withColumn(
    "appointment_date",
    to_date("appointment_date", "yyyy-MM-dd")
)
appointments.show()
```

```python
appointments = appointments.withColumn(
    "appt_month",
    month("appointment_date")
)
```

```python
from pyspark.sql.functions import count
appointments.groupBy("appointment_date") \
            .agg(count("*").alias("appointment_count")) \
            .show()
```

```python
appointments.groupBy("appt_month") \
            .agg(count("*").alias("appointment_count")) \
            .show()
```

```python
from pyspark.sql.functions import col
patients.filter(col("registration_date") > "2023-06-01").show()
```

```python
from pyspark.sql.functions import current_date, datediff
patients = patients.withColumn(
    "days_since_registration",
    datediff(current_date(), patients.registration_date)
)
patients.show()
```

```python
appointments = appointments.withColumn(
    "days_since_appointment",
    datediff(current_date(), appointments.appointment_date)
)
appointments.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import rank, row_number, dense_rank, col
window_spec = Window.partitionBy("city").orderBy(col("age").desc())
patients = patients.withColumn(
    "age_rank",
    rank().over(window_spec)
)
patients.show()
```

```python
window_spec = Window.partitionBy("city").orderBy(col("age").desc())
oldest_patients = patients.withColumn(
    "row_num",
    row_number().over(window_spec)
).filter(col("row_num") == 1)

oldest_patients.show()
```

```python
window_spec = Window.partitionBy("specialization").orderBy(col("consultation_fee").desc())
doctors = doctors.withColumn(
    "fee_rank",
    dense_rank().over(window_spec)
)
doctors.show()
```

```python
window_spec = Window.partitionBy("specialization").orderBy(col("consultation_fee").desc())
top_doctors = doctors.withColumn(
    "row_num",
    row_number().over(window_spec)
).filter(col("row_num") == 1)
top_doctors.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, col
window_spec = Window.partitionBy("city").orderBy(col("age").desc())
top2 = patients.withColumn("rn", row_number().over(window_spec)) \
               .filter(col("rn") <= 2)
top2.show()
```

```python
from pyspark.sql.functions import rank
window_spec = Window.orderBy(col("consultation_fee").desc())
doctors = doctors.withColumn(
    "fee_rank",
    rank().over(window_spec)
)
doctors.show()
```

```python
from pyspark.sql.functions import count
city_count = patients.groupBy("city").agg(count("*").alias("patient_count"))
window_spec = Window.orderBy(col("patient_count").desc())
city_rank = city_count.withColumn(
    "city_rank",
    rank().over(window_spec)
)
city_rank.show()
```

```python
appt_count = appointments.groupBy("doctor_id") \
                         .agg(count("*").alias("appt_count"))
window_spec = Window.orderBy(col("appt_count").desc())
doctor_rank = appt_count.withColumn(
    "rank",
    rank().over(window_spec)
)
doctor_rank.show()
```

```python
from pyspark.sql.functions import sum
window_spec = Window.partitionBy("payment_mode") \
                    .orderBy("bill_id") \
                    .rowsBetween(Window.unboundedPreceding, 0)
bills = bills.withColumn(
    "running_total",
    sum("bill_amount").over(window_spec)
)
bills.show()
```

```python
window_spec = Window.partitionBy("doctor_id") \
                    .orderBy("appointment_date") \
                    .rowsBetween(Window.unboundedPreceding, 0)
appointments = appointments.withColumn(
    "running_count",
    count("*").over(window_spec)
)
appointments.show()
```
