```python
from pyspark.sql.functions import when
accounts.withColumn("balance_category",
    when(col("balance") >= 100000, "High")
    .when(col("balance") >= 50000, "Medium")
    .otherwise("Low")
).show()
```

```python
customers.withColumn("age_group",
    when(col("age") < 30, "Young")
    .when(col("age") < 40, "Adult")
    .otherwise("Senior")
).show()
```

```python
transactions.withColumn("txn_category",
    when(col("amount") >= 30000, "Large")
    .when(col("amount") >= 10000, "Medium")
    .otherwise("Small")
).show()
```

```python
branches.withColumn("priority",
    when(col("region") == "South", "High Priority")
    .when(col("region") == "North", "Medium Priority")
    .otherwise("Watch")
).show()
```

```python
customers.withColumn("type_label",
    when(col("account_type") == "Savings", "Retail")
    .otherwise("Business")
).show()
```

```python
from pyspark.sql.window import Window

w_city = Window.partitionBy("city").orderBy(col("balance").desc())
```

```python
cust_acc.withColumn("rank", rank().over(w_city)).show()
```

```python
cust_acc.withColumn("row_num", row_number().over(w_city)).show()
```

```python
w_txn = Window.partitionBy("txn_type").orderBy(col("amount").desc())
transactions.withColumn("dense_rank", dense_rank().over(w_txn)).show()
```

```python
w_acc = Window.partitionBy("account_id").orderBy(col("amount").desc())
transactions.withColumn("rank", rank().over(w_acc)).filter("rank <= 2").show()
```

```python
w_acc = Window.partitionBy("account_id").orderBy(col("amount").desc())
transactions.withColumn("rank", rank().over(w_acc)).filter("rank <= 2").show()


```

```python
w_run = Window.partitionBy("account_id").orderBy("txn_date")
transactions.withColumn("running_total", sum("amount").over(w_run)).show()

```
