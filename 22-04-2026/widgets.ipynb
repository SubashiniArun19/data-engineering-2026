```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("PySpark Practice 2").getOrCreate()
print("Spark Session Created")
```

```python
data = [
    (1, "Rahul", "IT", 70000, "2024-01-10"),
    (2, "Sneha", "HR", 60000, "2024-02-15"),
    (3, "Arjun", "IT", 75000, "2024-03-01"),
    (4, "Priya", "Finance", 80000, "2024-01-25"),
    (5, "Karan", None, 50000, "2024-02-20")
]

columns = ["emp_id", "name", "department", "salary", "joining_date"]

df = spark.createDataFrame(data, columns)
df.show()
```

```python
from pyspark.sql.functions import to_date,year,month,dayofmonth
df_dates=df.withColumn("joining_date",to_date("joining_date"))
df_dates.show()
```

```python
df_dates.select("name","joining_date",
                year("joining_date").alias("year"),
                month("joining_date").alias("month"),
                dayofmonth("joining_date").alias("day")).show()
```

```python
new_data=[(6,"Meera","IT",72000,"2024-04-01"),
          (7,"Amit","HR",58000,"2024-04-10")]
new_df=spark.createDataFrame(new_data,columns)
new_df.show()
```

```python
df.union(new_df).show()
```

```python
df.unionByName(new_df).show()
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
def salary_band(salary):
  if salary >=75000:
    return "High"
  elif salary >=60000:
    return "Medium"
  else:
    return "Low"
salary_band_udf=udf(salary_band,StringType())
df.withColumn("salary_band",salary_band_udf("salary")).show()
```

```python
import ipywidgets as widgets
from IPython.display import display
```

```python
name_widget=widgets.Text(value="Subashini",description="Name:")
display(name_widget)
```

```python
dept_widget=widgets.Dropdown(options=["IT","HR","Finance"],value="IT",description="Dept: ")
display(dept_widget)
```

```python
salary_silder=widgets.IntSlider(value=60000,min=30000,max=100000,step=5000,description="Salary:")
display(salary_silder)
```

```python
button=widgets.Button(description="Show Values")
def on_button_Click(b):
  print("Name:",name_widget.value)
  print("Department:",dept_widget.value)
  print("Salary:",salary_silder.value)
button.on_click(on_button_Click)
display(button)
```
